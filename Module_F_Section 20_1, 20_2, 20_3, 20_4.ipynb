#Numerical Differentiation Problem Statement

Continuous functions can be converted to discrete functions if its values are only known at certain discreet locations in order to run numerical calculations relating to the function. Using this discretized function, the derivative can be found approximately.

# Finite Difference Approximating Derivatives

There are three ways to approximate the derivative using discrete values. **Forward difference** uses a point with a smaller x value than the desired point and finds the slope between it and the desired point. **Backward difference** uses a point with a larger x value than the desired point and finds the slope between it and the desired point. **Central difference** uses points with x values on either side of the desired point and uses the slope between the two points to approximate the slope at the desired point.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

# step size
h = 0.1
# define grid
x = np.arange(0, 2*np.pi, h) 
# compute function
y = np.cos(x) 

# compute vector of forward differences
forward_diff = np.diff(y)/h 
# compute corresponding grid
x_diff = x[:-1:] 
# compute exact solution
exact_solution = -np.sin(x_diff) 

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x_diff, forward_diff, '--', \
         label = 'Finite difference approximation')
plt.plot(x_diff, exact_solution, \
         label = 'Exact solution')
plt.legend()
plt.show()

# Compute max error between 
# numerical derivative and exact solution
max_error = max(abs(exact_solution - forward_diff))
print(max_error)
```

# Numerical Differentiation with Noise

When using data, the functions are often not perfect and are instead contaminated with **noise**.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

x = np.arange(0, 2*np.pi, 0.01) 
# compute function
omega = 100
epsilon = 0.01

y = np.cos(x)**2 
y_noise = y + epsilon*np.sin(omega*x)

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', \
         label = '$cos^2(x)$ + noise')
plt.plot(x, y, 'b-', \
         label = '$cos^2(x)$')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

x = np.arange(0, 2*np.pi, 0.01) 
# compute function
y = -2*np.cos(x)*np.sin(x) 
y_noise = y + epsilon*omega*np.cos(omega*x)

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', \
         label = 'Derivative $cos^2(x)$ + noise')
plt.plot(x, y, 'b-', \
         label = 'Derivative of $cos^2(x)$')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```
