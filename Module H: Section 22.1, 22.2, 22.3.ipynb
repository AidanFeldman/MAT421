#22.1 ODE Initial Value Problem Statement
An **Ordinary Differential Equation** is a derivative where the independent variable is in one dimension. Many practical situations are such that there is a known initial value with the change measured with respect to another variable (usually time).

#22.2 Reduction of Order
There are many numerical methods for solving differential equations designed specifically for first order ODEs. Therefore, it is useful to be able to reduce higher level ODEs to first order. A state of a system is a measure of the equations relevant to a system at any given time, including a base function and all of the pertinent derivatives. Then take the derivative of the state to obtain a first order ODE.

# 22.3 The Euler Method
The Euler Method uses a linear approximation of some function that represents the derivative of the given function. Starting from a point, x, where the derivative is already known, it is possible to estimate the function at some value x+h by using the derivative at x as an estimate of how the function behaves over the entire interval [x, x+h]. Under that assumption, the point x+h is approximately x plus the product of h and the derivative at x.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

f = lambda t, s: np.cos(t) # This is the given ODE
h = 0.1 # This is the step size we are using. The lower the step size, the lower the error.
t = np.arange(0, 1 + h, h)
s0 = 0 # The given value.

# Euler's method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, np.sin(t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
